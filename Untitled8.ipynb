```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Wine dataset
wine_data = pd.read_csv(r"C:\Users\Laptop\Downloads\PCA\PCA\wine.csv")

# Task 1: Basic data exploration
print("First 5 rows of the dataset:")
print(wine_data.head())

print("\nDataset information:")
print(wine_data.info())

# Task 2: Examine the distribution of features
# Histograms
wine_data.hist(figsize=(12, 10))
plt.suptitle('Histograms of Wine Dataset Features', x=0.5, y=1.02, ha='center', fontsize='x-large')
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()

# Box plots
plt.figure(figsize=(12, 8))
sns.boxplot(data=wine_data)
plt.title('Box plots of Wine Dataset Features')
plt.xticks(rotation=45)
plt.show()

# Task 3: Investigate correlations between features
correlation_matrix = wine_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Wine Dataset Features')
plt.show()
```

```python
####Task 3: Clustering with Original Data:
####1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
####2.	Visualize the clustering results using appropriate plots.
####3.	Evaluate the clustering performance using metrics such as silhouette score or Daviesâ€“Bouldin index.
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np

# Fit K-means clustering algorithm to the original dataset
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(wine_data)

# Visualize clustering results
# Since the dataset has many features, we'll reduce dimensionality using PCA for visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
wine_pca = pca.fit_transform(wine_data)

plt.figure(figsize=(10, 6))
plt.scatter(wine_pca[:, 0], wine_pca[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=100)
plt.title('K-means Clustering of Wine Dataset (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()

# Evaluate clustering performance
silhouette_score_value = silhouette_score(wine_data, clusters)
davies_bouldin_score_value = davies_bouldin_score(wine_data, clusters)

print("Silhouette Score:", silhouette_score_value)
print("Davies-Bouldin Index:", davies_bouldin_score_value)
```

```python
#### Task 4: Clustering with PCA Data:
####1.	Apply the same clustering algorithm to the PCA-transformed dataset.
####2.	Visualize the clustering results obtained from PCA-transformed data.
####3.	Compare the clustering results from PCA-transformed data with those from the original dataset.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load the dataset
wine_data = pd.read_csv(r"C:\Users\Laptop\Downloads\PCA\PCA\wine.csv")

# Separate features and labels
X = wine_data.drop('Type', axis=1)
y = wine_data['Type']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply KMeans clustering on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
y_kmeans_pca = kmeans_pca.fit_predict(X_pca)

# Visualize clustering results from PCA-transformed data
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans_pca, cmap='viridis', s=50, alpha=0.5)
plt.title('Clustering Results from PCA-transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()

# Calculate silhouette score for PCA-transformed data clustering
silhouette_avg_pca = silhouette_score(X_pca, y_kmeans_pca)
print(f"Silhouette Score for PCA-transformed Data: {silhouette_avg_pca}")

# Compare clustering results from PCA-transformed data with original data
silhouette_avg_original = silhouette_score(X_scaled, y)
print(f"Silhouette Score for Original Data: {silhouette_avg_original}")
```

```python
####Task 5: Comparison and Analysis:
####1.	Compare the clustering sults obtained from the original dataset and PCA-transformed data.
####2.	Discuss any similarities or differences observed in the clustering results.
####3.	Reflect on the impact of dimensionality reduction on clustering performance.
####4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.
####-Compare clustering results from the original dataset and PCA-transformed data.
####Discuss similarities, differences, impact of dimensionality reduction, and trade-offs between using PCA and clustering directly on the original dataset.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load the dataset
wine_data = pd.read_csv(r"C:\Users\Laptop\Downloads\PCA\PCA\wine.csv")

# Separate features and labels
X = wine_data.drop('Type', axis=1)
y = wine_data['Type']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply KMeans clustering on original dataset
kmeans_original = KMeans(n_clusters=3, random_state=42)
y_kmeans_original = kmeans_original.fit_predict(X_scaled)

# Apply KMeans clustering on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
y_kmeans_pca = kmeans_pca.fit_predict(X_pca)

# Compare clustering results from original dataset and PCA-transformed data
silhouette_avg_original = silhouette_score(X_scaled, y_kmeans_original)
silhouette_avg_pca = silhouette_score(X_pca, y_kmeans_pca)

print("Silhouette Score for Original Data:", silhouette_avg_original)
print("Silhouette Score for PCA-transformed Data:", silhouette_avg_pca)

# Plotting the clustering results
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans_original, cmap='viridis', s=50, alpha=0.5)
plt.title('Clustering Results from Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')

plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans_pca, cmap='viridis', s=50, alpha=0.5)
plt.title('Clustering Results from PCA-transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')

plt.tight_layout()
plt.show()

# Discussion
print("\nComparison and Analysis:")
print("1. Similarities and Differences:")
print("   - Both methods aim to group data points into clusters.")
print("   - Differences arise due to dimensionality reduction in PCA-transformed data.")
print("2. Impact of Dimensionality Reduction:")
print("   - PCA reduces complexity and noise, potentially improving clustering performance.")
print("   - However, it may also lead to information loss.")
print("3. Trade-offs:")
print("   - PCA reduces computational complexity and focuses on significant features.")
print("   - Clustering on original data preserves all features but may suffer from high dimensionality.")
```
